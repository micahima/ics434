#### 1. Reproducing a plot
1. Use `numpy`'s `random` module to generate a random list of size 1000 of the following three categories: "CS", "Engineering", "Math" and "Physics"
  1.1 Use a seed of 42
2. Store the data in an array called my `students_affiliations`
3. use `matplotlib` to plot the data in `students_affiliations` to reproduce the following graph
 3.1 The plot uses "#FF5733" as color with an alpha (transparency) of 0.55.

* Your graph should look exactly like the one below

![](https://www.dropbox.com/s/ffmu62es6d7evh5/experience_fig.png?dl=1)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
disciplines = [ "CS", "Engineering", "Math", "Physics"]
students_affiliations = pd.Series(np.random.choice(disciplines, size=1000))
counts = [students_affiliations.value_counts()["CS"], 
          students_affiliations.value_counts()["Engineering"], 
          students_affiliations.value_counts()["Math"],
          students_affiliations.value_counts()["Physics"]]
df = pd.DataFrame({'Counts': counts, 'Disciplines': disciplines})
sorted_df = df.sort_values(by='Counts', ascending=False)
```

```python
fig = plt.figure()
ax = fig.add_subplot() 
ax.bar(sorted_df['Disciplines'], sorted_df['Counts'], color='#FF5733', alpha=0.55)
ax.set_title("Distribution of Students in the Program", color='green')
ax.set_xlabel("Disciplines")
ax.set_ylabel("Counts")
```

* Generate random data to produce a scatter plot

```python
age = np.random.randint(20, 45, 100)
takeoutExpense = np.random.randint(100, 500, 100)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(age, takeoutExpense)

plt.xlabel("Age")
plt.ylabel("Amount Spent on Takeout")
plt.title("Representation of participants Age and Monthly Spending on Takeout")
```
